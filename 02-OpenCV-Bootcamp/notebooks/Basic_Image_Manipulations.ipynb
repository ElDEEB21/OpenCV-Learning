```python
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

from IPython.display import Image
```

```python
def download_and_unzip(url, save_path):
    print(f"Downloading and extracting assests....", end="")

    # Downloading zip file using urllib package.
    urlretrieve(url, save_path)

    try:
        # Extracting zip file using the zipfile package.
        with ZipFile(save_path) as z:
            # Extract ZIP file contents in the same directory.
            z.extractall(os.path.split(save_path)[0])

        print("Done")

    except Exception as e:
        print("\nInvalid file.", e)
URL = r"https://www.dropbox.com/s/rys6f1vprily2bg/opencv_bootcamp_assets_NB2.zip?dl=1"

asset_zip_path = os.path.join(os.getcwd(), f"opencv_bootcamp_assets_NB2.zip")

# Download if assest ZIP does not exists. 
if not os.path.exists(asset_zip_path):
    download_and_unzip(URL, asset_zip_path)   
```

```python
img_NZ_bgr = cv.imread('../Photos/New_Zealand_Lake.jpg')
plt.imshow(img_NZ_bgr[..., ::-1])  # Convert BGR to RGB for display
```

```python
img_NZ_rgb = img_NZ_bgr[..., ::-1]  # Convert BGR to RGB
cropped_img = img_NZ_rgb[100:400, 100:400]  # Crop a region of interest
plt.imshow(cropped_img)
```

```python
resized_cropped_img = cv.resize(cropped_img, (200, 200), interpolation=cv.INTER_AREA)
plt.imshow(resized_cropped_img)
```

```python
img_NZ_rgb_flipped_hor = cv.flip(img_NZ_rgb, 1)  # Flip horizontally
img_NZ_rgb_flipped_ver = cv.flip(img_NZ_rgb, 0)  # Flip vertically
img_NZ_rgb_flipped_both = cv.flip(img_NZ_rgb, -1)  # Flip both horizontally and vertically

plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1)
plt.imshow(img_NZ_rgb_flipped_hor)
plt.title('Flipped Horizontally')
plt.subplot(1, 4, 2)
plt.imshow(img_NZ_rgb_flipped_ver)
plt.title('Flipped Vertically')
plt.subplot(1, 4, 3)
plt.imshow(img_NZ_rgb_flipped_both)
plt.title('Flipped Both')
plt.subplot(1, 4, 4)
plt.imshow(img_NZ_rgb)
plt.title('Original Image')
plt.tight_layout()
plt.show()
```
